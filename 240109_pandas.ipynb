## GroupBy 연산
- SQL : Group By와 Logic은 동일함
- Groupby 기준 컬럼: 패턴이 있는 문자열
- Groupby 대상 컬럼: 수치데이터
- 필요한 함수: 집계함수 (합, 평균, 표준편차, 최대, 최소 등)

```python
# df.groupby('기준컬럼')['대상컬럼']

df.groupby('Location')['매출액'].agg(['sum', 'mean', 'std'])
```

```python
df.groupby('Location')['매출액'].agg(['sum', 'mean', 'std']).reset_index()
```
