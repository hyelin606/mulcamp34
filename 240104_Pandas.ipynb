```python
# df_KTX.loc[행 조건식, [컬럼명]]

iris = sns.load_dataset('iris')


# sepal_length의 길이가 7cm 이상인 것만 조회
iris.loc[iris['sepal_length'] >=7, :]
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# embark_town 컬럼에서 Southampton 거주민의 모든 데이터를 조회

titanic.loc[titanic['embark_town'] == 'Southampton' , :]
```

```python
# age 평균 구하고, 평균 이상인 데이터 조회

titanic.loc[titanic['age'] >= titanic['age'].mean(), :]
```

```python
var = 'fare'
titanic.loc[titanic[var] >= titanic[var].mean(), :].head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# total_bill의 중간값 이하인 것만 조회
# 중간값이란 서열을 매기면 50번째 값
```

```python
tips['total_bill'].median()
```

```python
tips.loc[tips['total_bill'] <= tips['total_bill'].median(), :]
```

```python
# day가 Sun인 데이터만 조회히기
```

```python
tips.loc[tips['day'] == 'Sun' , :]
```

```python
# tip의 평균보다 크면서, time 이 dinner인 데이터만 조회
# AND 조건, A이고 B이다. A이면서 B이다.
# OR 조건, A이거나 B이다.

# &, AND 조건 / |, OR 조건

#tips.loc[(조건1) & (조건2), :]

tips.loc[(tips['tip'] >= tips['tip'].mean()) & (tips['time'] == 'Dinner'), :]
```

```python
# 다중조건 예시 (조건을 계속 추가할 수 있음)

tips.loc[(tips['tip'] >= tips['tip'].mean()) &
    (tips['time'] == 'Dinner') &
    (tips['size'] == 3) &
    (tips['sex'] == 'Male') &
    (tips['day'] == 'Sat'),
:]
```
